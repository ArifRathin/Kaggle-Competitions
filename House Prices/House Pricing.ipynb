```python
train_data.drop(['Alley','MasVnrType','FireplaceQu','PoolQC','Fence','MiscFeature'], axis=1, inplace=True)
train_data.isnull().sum()
```

```python
cols_to_drop = ['Alley','MasVnrType','FireplaceQu','PoolQC','Fence','MiscFeature']
test_data.drop(cols_to_drop, axis=1, inplace=True)
test_data.isnull().sum()
```

```python
##TO Numerize:MSZoning, Street, Alley, LotShape, LandContour, Utilities, LotConfig, LandSlope, Neighborhood, Condition1, Condition2, BldgType, HouseStyle,
##RoofStyle, RoofMatl, Exterior1st, Exterior2nd, MasVnrType, ExterQual, ExterCond, Foundation, BsmtQual, BsmtCond, BsmtExposure, BsmtFinType1, BsmtFinType2
##Heating, HeatingQC, CentralAir, Electrical, KitchenQual, Functional, FireplaceQu, GarageType, GarageFinish, GarageQual, GarageCond, PavedDrive
##PoolQC, Fence, MiscFeature, SaleType, SaleCondition
```

```python
cols_to_numerize = ['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle',
                   'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
                   'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive',
                    'PoolQC', 'Fence', 'MiscFeature', 'SaleType', 'SaleCondition'
                   ]
cols_to_numerize = list(set(cols_to_numerize)-set(cols_to_drop))
cols_to_numerize
```

```python
for col in cols_to_numerize:
    train_data.fillna({col:train_data[col].mode()[0]}, inplace=True)
    test_data.fillna({col:test_data[col].mode()[0]}, inplace=True)
train_data.head()
```

```python
train_data.isnull().sum()
```

```python
test_data.isnull().sum()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
for col in cols_to_numerize:
    train_data[col] = label_encoder.fit_transform(train_data[col])
    test_data[col] = label_encoder.transform(test_data[col])
```

```python
train_data.info()
```

```python
train_data.head(10)
```

```python
train_cols_with_na = list(train_data.columns[train_data.isnull().any()])
train_cols_with_na
```

```python
for col in train_cols_with_na:
    train_data.fillna({col:train_data[col].mean()},inplace=True)
train_data.isnull().sum()
```

```python
test_cols_with_na = list(test_data.columns[test_data.isnull().any()])
test_cols_with_na
```

```python
for col in test_cols_with_na:
    test_data.fillna({col:test_data[col].mean()},inplace=True)
test_data.isnull().sum()
```

```python
Y_train = train_data['SalePrice']
X_train = train_data.drop('SalePrice',axis=1)
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
linear_regression.fit(X_train, Y_train)
prediction = linear_regression.predict(test_data)
prediction
```
