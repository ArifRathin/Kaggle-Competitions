```python
import pandas as pd
```

```python
train_data = pd.read_csv(r"F:\Machine Learning\Kaggle Projects\Spaceship Titanic\spaceship-titanic\train.csv")
full_test_data = pd.read_csv(r"F:\Machine Learning\Kaggle Projects\Spaceship Titanic\spaceship-titanic\test.csv")
passenger_id = full_test_data['PassengerId']
train_data.head()
```

```python
train_data.shape
```

```python
train_data.isnull().sum()
```

```python
train_data.info()
```

```python
train_data.insert(4,'Deck',None)
train_data.insert(5,'Side',None)
train_data.head()
```

```python
train_data.fillna({'Cabin':'*/*/*'},inplace=True)
```

```python
train_data.Cabin.isnull().sum()
```

```python
import numpy as np
```

```python
i=0
for cabin in train_data['Cabin']:
    if cabin.split('/')[0] == '*':
        train_data.loc[i,'Deck'] = np.nan
        train_data.loc[i,'Side'] = np.nan
    else:
        train_data.loc[i,'Deck'] = cabin.split('/')[0]
        train_data.loc[i,'Side'] = cabin.split('/')[2]
    i += 1
train_data.head()
```

```python
train_data.isnull().sum()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
train_data['HomePlanet'] = le.fit_transform(train_data['HomePlanet'])
print(le.classes_)
train_data['CryoSleep'] = le.fit_transform(train_data['CryoSleep'])
print(le.classes_)
train_data['Deck'] = le.fit_transform(train_data['Deck'])
print(le.classes_)
train_data['Side'] = le.fit_transform(train_data['Side'])
print(le.classes_)
train_data['Destination'] = le.fit_transform(train_data['Destination'])
print(le.classes_)
train_data['VIP'] = le.fit_transform(train_data['VIP'])
print(le.classes_)
train_data['Transported'] = le.fit_transform(train_data['Transported'])
print(le.classes_)
```

```python
train_data.drop(['PassengerId','Cabin','Name'],axis=1,inplace=True)
```

```python
import math
train_data.fillna({'Age':math.floor(train_data['Age'].mean())},inplace=True)
train_data.fillna({'RoomService':math.floor(train_data['RoomService'].mean())},inplace=True)
train_data.fillna({'FoodCourt':math.floor(train_data['FoodCourt'].mean())},inplace=True)
train_data.fillna({'ShoppingMall':math.floor(train_data['ShoppingMall'].mean())},inplace=True)
train_data.fillna({'Spa':math.floor(train_data['Spa'].mean())},inplace=True)
train_data.fillna({'VRDeck':math.floor(train_data['VRDeck'].mean())},inplace=True)
```

```python
train_data.head()
```

```python
train_data.insert(12,'total_spent',0)
train_data.head()
```

```python
i = 0
total = 0
for idx,td in train_data.iterrows():
    total = td['RoomService']+td['FoodCourt']+td['ShoppingMall']+td['Spa']+td['VRDeck']
    train_data.loc[i,'total_spent'] = total
    i += 1
```

```python
train_data.drop(['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck'],axis=1,inplace=True)
train_data.head()
```

```python
mean_total_spent = math.floor(train_data['total_spent'].mean())
```

```python
train_data.insert(8,'grt_mean_spent',0)
train_data.loc[train_data['total_spent'] > mean_total_spent,'grt_mean_spent'] = 1
train_data.head()
```

```python
train_data.drop(['total_spent'],axis=1,inplace=True)
mean_age = math.floor(train_data['Age'].mean())
mean_age
```

```python
train_data.insert(6,'grt_mean_age',0)
train_data.loc[train_data['Age']>mean_age,'grt_mean_age'] = 1
train_data.drop(['Age'],axis=1,inplace=True)
train_data.head()
```

```python
train_data.isnull().sum()
```

```python
test_data = pd.read_csv(r"F:\Machine Learning\Kaggle Projects\Spaceship Titanic\spaceship-titanic\test_processed.csv")
test_data.head()
```

```python
y_train = train_data['Transported']
X_train = train_data.drop(['Transported'],axis=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier(n_neighbors = 40)
kn.fit(X_train, y_train)
predictions = kn.predict(test_data)
predictions
```

```python
y_true = pd.read_csv(r"F:\Machine Learning\Kaggle Projects\Spaceship Titanic\spaceship-titanic\sample_submission.csv")
y_true.drop(['PassengerId'],axis=1,inplace=True)
y_true.head()
```

```python
y_true['Transported'] = le.fit_transform(y_true['Transported'])
print(le.classes_)
```

```python
print(f"Accuracy: {kn.score(test_data, y_true)}")
```
